<h2>Let's make part 2 more exciting with a little data visualization!</h2>

<b>Next up is the plotting of the data. I hope to make more posts on visualizations and how they work, but in the meantime, just copy and paste the code I have and you'll see that the data plots nicely, thanks to the index.</b>

```python
df.plot()
```

You can see how easy it was to plot the graph and visualize the data. It really helped that the index was setup and it allowed us to easily plot the data as a timeline.

<b>Let's try getting rid of the other aspects of the dataset and only leaving the 'Open' and 'Close' columns. It gives us an idea of how much the value of the coin changes within a day.</b>

```python
df[['Open', 'Close']].plot()
```

As expected, from what I've seen of Bitcoin over the past little while, it opens and closes at around the same mark.

<b>You're probably wondering, "that's all cool and well, but how do we focus on 2017 data?" Really simple! To get only the data for 2017 all you have to do is the following...</b>

```python
df2017 = df[df.index.year == 2017]
```

We already used this function above to get a sense of the first 5 rows (in <a href="https://datadatagoose.wordpress.com/2018/01/09/beginner-getting-comfortable-with-data-using-bitcoin-market-data-part-1-python/">part 1</a>), but let's use it again in the new dataset which is limited to 2017. If you wish to get more rows, you can always specify that number in between the parentheses (e.g. <i>df2017.head(10)</i> for the first 10 rows). You could always look at the last 5 rows of the dataset by using the tail function instead of the head function (i.e. <i>df2017.tail()</i> )

```python
df2017.head()
```

<b>Visaulization time! I've plotted the 'High' for each day in 2017.</b>

```python
df2017['High'].plot()
```

Wow! 2017 was a roller coaster ride for Bitcoin! Plot the other columns to get some practice in.

Below you'll find Bitcoin's market price difference between the days (in USD).

```python
df2017.diff().head()
```

I thought we were going to see the difference between everyday, then why is the first row showing "NaN" across the board? What does "NaN" even mean?

<b>"NaN" stands for "Not a Number". The reason this shows up for January 1st, 2017, is because their is no previous day in our dataset in order for it to calculate a difference, hence the "NaN".</b>

Cool. So far so good. We're almost at the end of our 2 part series of how I get comfortable with my data.

<b>Usually, along with seeing the difference, as data people, we're more interested in the change in percentage. Brace yourselves, because the following graph shows how much of a roller coaster ride 2017 was for Bitcoin!</b>

```python
(df2017['Open'].pct_change()*100).plot()
```

Woah! That visualization tells quite the story! Look at how steep those changes are! No wonder this thing made it all over the news.

<b>Lastly, I'm going to call upon a function by the name of "describe". What this function does is gives us a very high level overview of our dataset.</b> I used it to give us a high level overview of our change in percentage. Let's take a look.

```python
(df2017['Open'].pct_change()*100).describe()
```

Cool! So much insight in such little code. That's very powerful!

<b>What we're seeing is in 365 days (the count is showing 364, because remember the first day is "NaN"), the highest rise in percentage change was 23% in a single day and the lowest was 17%. We also see that the average (mean) change in percentage was 0.8%.</b>

There's a few other things we can get from this "describe" function, but I think we covered a lot in this 2 part series.
